```python
data = [
  {
    "id":1,
    "name":"cole volk",
    "fitness":{"height":130,"weight":60}
  },
  {
    "name":"mark",
    "fitness":{"height":140,"weight":80}
  },
  {
    "id":2,
    "name":"Ragner",
    "fitness":{"height":150,"weight":70}
  },

]
```

```python
pd.json_normalize(data)
```

```python
#only shows the top level data
pd.json_normalize(data, max_level=0)
```

```python
#going to one more nested level
pd.json_normalize(data, max_level=1)
```

```python
data = [
    {
        "state": "Florida",
        "shortname": "FL",
        "info": {"governor": "Rick Scott"},
        "counties": [
            {"name": "Dade", "population": 12345},
            {"name": "Broward", "population": 40000},
            {"name": "Palm Beach", "population": 60000},
        ],
    },
    {
        "state": "Ohio",
        "shortname": "OH",
        "info": {"governor": "John Kasich"},
        "counties": [
            {"name": "Summit", "population": 1234},
            {"name": "Cuyahoga", "population": 1337},
        ],
    },
]
```

```python
#it will not recognise inner list as seperate key value pair dictionary
pd.json_normalize(data)
```

```python
pd.json_normalize(data,"counties")
```

```python
pd.json_normalize(data,"counties",["state",'shortname'])
```

```python
#for getting governor value also
pd.json_normalize(data,"counties",["state",'shortname',['info','governor']])
```
