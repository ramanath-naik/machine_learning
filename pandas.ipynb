```python
#Series
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
print(myvar[0])
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

var = pd.Series(calories)

print(var)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

var2 = pd.Series(calories, index = ["day1", "day2"])

print(var2)
```
